## Background

**Problem**

Tourism was hit hard by the COVID-19 pandemic. Being a major key contributor for the country's GDP, Portugal's National Tourism Board Organizations (NTBO) launched in May 2021 the *Reactivate Tourism | Build the Future Action Plan program*, which aims to be a guiding script for the national tourism sector recovery, public and private, from the pandemic. 

As part of this program, NTBO wants to start studying visitors' generated content, such as TripAdvisory users reviews, to understand visitors' behavior pattern and preferences anf if these were affected by the pandemic.

NTBO consulted us to characterize and describe the patterns of visitants of Portuguese attractions and comparing it to Portugal's main tourism competitors by employing CRISP-DM model.

**Background: Tourism in Portugal in recent years**

According to the most recent Tourism Satellite Account from the National Statistics Institute (INE), in 2022* the Gross Value Added generated by Tourism (GVAT) totalled 18.5 million euros, a nominal increase of 72.7%. GVAT represented 8.9% of national GVA (Graph 1), more 3.2 percentage points  than in 2021, surpassing pre-pandemic levels (in 2019 GVAT represented 8.1% of the economy's GVA).

**2022 values are preliminary data.*


```python
import matplotlib.pyplot as plt

# Creating a bar plot to see GVAGT Evolution in last 4 years
categories = ['2019', '2020', '2021', '2022']
values = [8.1, 4.4, 5.7, 8.9]

bars = plt.bar(categories, values, color='salmon')

# Adding labels and title
plt.xlabel('Years')
plt.ylabel('GVAT (%)')
plt.title('1. GVAT weight evolution in Portuguese GVA (2019-2022)')

# Display values on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom')
    
# Showing the plot
plt.show()
```

In 2022 Tourism Consumption in the Economic Territory (TCET) was equivalent to **15.8 per cent of GDP** (Graph 2), 6 percentage points more than in 2021, also above 2019 levels, when it was equivalent to 15.3 per cent of GDP. 

Tourism is estimated to have generated a direct and indirect contribution of **29.2 billion euros to GDP in 2022**, which corresponds to 12.2 per cent (7.8 per cent in 2021 and 6.6 per cent in 2020). Overall, 2022 GVAT and TCET values were above 2019 levels, showing a more marked recovery dynamic to the pre-pandemic levels than that observed for the total national economy in 2022.

**Main effects of the COVID-19 pandemic on tourism in 2020**

In 2020, total GVAT **declined by 48.9% compared to 2019**, while the national economy's GVA fell by 5.8% in the same year (Graph 3 for a more comprehensive look between variables and years). According to INE, there was also a sharper decline in the GVAT of tourism-related activities (-49.0%).

When it comes to TCET values, the figure is much similar: tourism consumer in total economy declined by **48.8%** in 2020.

```python
# Creating a combined plot to see TCET and GVAT Evolution pre and post pandemic

import matplotlib.pyplot as plt
import numpy as np

categories = ['2019', '2020', '2021', '2022']
values1 = [32.9, 16.8, 21.1, 37.8]
values2 = [15.1, 7.7, 10.6, 18.3]
values3 = [15.3, 8.4, 9.8, 15.8]
values4 = [8.1, 4.4, 5.7, 8.9]

fig, ax1 = plt.subplots()
bar_width = 0.2
bar1 = ax1.bar(np.arange(len(categories)) - bar_width, values1, bar_width, label='TCET', color='blue')
bar2 = ax1.bar(np.arange(len(categories)), values2, bar_width, label='GVAT', color='yellowgreen')

# Create a line plot for the % variables
ax2 = ax1.twinx()
ax2.plot(categories, values3, marker='o', linestyle='-', color='darkblue', label='TCET weight in GDP (%)')
ax3 = ax1.twinx()
ax3.plot(categories, values4, marker='s', linestyle='-', color='darkgreen', label='GVAT weight in GDP (%)')

# Add title
ax1.set_title('TCET and GVAT evolution pre and post COVID-19')

# Display legends
ax1.legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), fancybox=True, shadow=True, ncol=3)
ax2.legend(loc='upper center', bbox_to_anchor=(0.5, -0.3), fancybox=True, shadow=True, ncol=3)
ax3.legend(loc='upper center', bbox_to_anchor=(0.5, -0.4), fancybox=True, shadow=True, ncol=3)


# Set limits for the secondary y-axes (percentages)
ax2.set_ylim(0, 20)
ax3.set_ylim(0, 20)

# Show the plot
plt.show()
```

The decrease in domestic tourism expenditure (-34.7%), compared to inbound tourism expenditure (-59.4%), resulted in an increase in the weight of domestic tourism in the total TCET, which rose from 30.6% in 2019 to 39.0% in 2020. Conversely, there was a reduction in the weight of inbound tourism expenditure (from 64.4% in 2019 to 51.0% in 2020), while still maintaining its primacy over domestic tourism expenditure and the other components.

## Business Objectives
Understanding the tourism business landscape in these last few years, and the available information to analyse, we defined the following business objectives for our analysis: 

**1. Understand Visitor Demographics:**

**Objective**: Analyse user-generated content to identify and understand the demographics of visitors to Portuguese attractions. 

**Rationale**: Knowing the nationality and other demographic factors of visitors helps tailor marketing strategies and services to better meet their preferences.

**2. Identify Popular Attractions and Activities:** 

**Objective**: Determine the most popular attractions and activities based on user-generated content. 

**Rationale**: Knowing which attractions and activities are highly rated and frequently mentioned by visitors helps in promoting key experiences and optimizing resource allocation.

**3. Assess Seasonal and Temporal Patterns:**

**Objective**: Analyse patterns in visitation based on temporal factors such as seasons, holidays, and special events. 

**Rationale**: Understanding when peak visitation occurs and any seasonal variations can inform strategic planning, resource allocation, and promotional efforts.

**4. Examine Impact of the COVID-19 Pandemic:**

**Objective**: Evaluate how visitor patterns have been affected by the COVID-19 pandemic. 

**Rationale**: Assessing changes in visitor behaviour and preferences in the post-pandemic era is crucial for adapting strategies and services to the new normal.

**5. Benchmark against Competitors:** 

**Objective**: Compare user-generated content and visitor patterns for Portuguese attractions with those of key competitors in the tourism industry. 

**Rationale**: Benchmarking against competitors provides insights into Portugal's competitive position, allowing the NTBO to identify strengths, weaknesses, opportunities, and threats.

**6. Evaluate Visitors Sentiments:**

**Objective**: Assess sentiment analysis of TripAdvisor reviews to gauge visitor satisfaction and identify areas for improvement. 

**Rationale**: Understanding the sentiment of reviews helps in identifying aspects of the visitor experience that are positively received or need attention, providing insights for enhancing tourism offerings.
